# **Fourier Transform and Frequency Spectrum Analysis of Signals**

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from scipy import signal
```

### **Q1. Sine wave**

```python
# a- to generate sine signal
Fs = 1000  # Sampling frequency - 1000 samples/sec
f = 5  # Frequency of wave-5 Hz
T = 2  # Duration- 2 second
t = np.linspace(0, T, int(Fs*T), endpoint=False)  # Time vector
signal = np.sin(2 * np.pi * f * t)
```

```python
# b- Plot the time-domain waveform of the signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.title('Time Domain Waveform of 5Hz Sine Wave')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# DTFT function
def dtft(signal, num_frequencies, Fs):
    N = len(signal)
    n = np.arange(N)
    omega = np.linspace(-np.pi, np.pi, num_frequencies)  # Frequency range in rad/s
    dtft_values = np.array([np.sum(signal * np.exp(-1j * w * n)) for w in omega])  # DTFT formula
    freqs = omega * Fs / (2 * np.pi)  # Convert rad/s to Hz
    return freqs, dtft_values

# c- Compute DTFT
num_frequencies = 1000  # Number of frequency points for the DTFT
dtft_frequencies, dtft_values = dtft(signal, num_frequencies, Fs)
```

```python
# Plot the DTFT continuous frequency spectrum
plt.figure(figsize=(10, 4))
plt.plot(dtft_frequencies, np.abs(dtft_values))
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
#d- computing dft
dft_values = np.fft.fft(signal)  # Compute the FFT
frequencies = np.fft.fftfreq(len(signal), d=1/Fs)  # Frequency bins

#to only take positive half of spectrum
pos_freq = frequencies[:len(frequencies)//2]
pos_dft = np.abs(dft_values[:len(frequencies)//2])
```

```python
# Plot the discrete frequency spectrum (DFT)
plt.figure(figsize=(10, 4))
plt.plot(pos_freq, pos_dft)
plt.title('Discrete Frequency Spectrum (DFT using FFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

### **Q2. Composit signal**- adding two sine waves

```python
# a- Composite signal- adding 2 sine waves
Fs = 1000  # Sampling freq
T = 1  # Duration -1 sec
t = np.linspace(0, T, int(Fs*T), endpoint=False)  # Time vector

# Sine wave with different freq and amplitude
f1 = 10   # Frequency 1st wave (5 Hz)
A1 = 2   # Amplitude 1st wave

f2 = 30  # Frequency 2nd wave (50 Hz)
A2 = 1   # Amplitude 2nd wave

# The composite signal -sum of two sine waves
signal2 = A1 * np.sin(2 * np.pi * f1 * t) + A2 * np.sin(2 * np.pi * f2 * t)
```

```python
# b- Plot the time-domain waveform of the composite signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal2)
plt.title('Time Domain Waveform of Composite Signal (10 Hz + 30 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# c- Compute DTFT function for the composite signal
num_frequencies = 1000  # Number of frequency points for the DTFT
dtft_frequencies, dtft_values = dtft(signal2, num_frequencies, Fs)
```

```python
# Plot the continuous frequency spectrum (DTFT)
plt.figure(figsize=(10, 4))
plt.plot(dtft_frequencies, np.abs(dtft_values))
plt.title('Continuous Frequency Spectrum (DTFT) of Composite Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
#d- computing dft
dft_values = np.fft.fft(signal2)  # Compute the FFT
frequencies = np.fft.fftfreq(len(signal2), d=1/Fs)  # Frequency bins

#to only take positive half of spectrum
pos_freq = frequencies[:len(frequencies)//2]
pos_dft = np.abs(dft_values[:len(frequencies)//2])
```

```python
# Plot the discrete frequency spectrum (DFT)
plt.figure(figsize=(10, 4))
plt.plot(pos_freq, pos_dft)
plt.title('Discrete Frequency Spectrum (DFT using FFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

### **Q3- Exponentially decaying signal.**

```python
#a- Exponentially signal
Fs = 1000  # Sampling freq
T = 1  # Duration - 1 second
t = np.linspace(0, T, int(Fs*T), endpoint=False)  # Time vector

# exponentially decaying signal
alpha = 5  # Decay rate
signal3 = np.exp(-alpha * t)
```

```python
# b- Plot the time-domain waveform of the exponentially decaying signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal3)
plt.title('Time Domain Waveform of Exponentially Decaying Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# c- Compute DTFT for the exponentially decaying signal
num_frequencies = 1000  # Number of frequency points for the DTFT
dtft_frequencies, dtft_values = dtft(signal3, num_frequencies, Fs)
```

```python
# Plot the continuous frequency spectrum (DTFT)
plt.figure(figsize=(10, 4))
plt.plot(dtft_frequencies, np.abs(dtft_values))
plt.title('Continuous Frequency Spectrum (DTFT) of Exponentially Decaying Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
#d- computing DFT
dft_values = np.fft.fft(signal3)
frequencies = np.fft.fftfreq(len(signal3), d=1/Fs)  # Frequency bins

# Only take the positive half of the frequency spectrum
pos_freq3 = frequencies[:len(frequencies)//2]
pos_dft3 = np.abs(dft_values[:len(frequencies)//2])
```

```python
# Plot the discrete frequency spectrum (DFT using FFT)
plt.figure(figsize=(10, 4))
plt.plot(pos_freq3, pos_dft3)
plt.title('Discrete Frequency Spectrum of Exponentially Decaying Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```
